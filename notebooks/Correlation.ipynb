```python
import pandas as pd
import yfinance as yf
import numpy as np
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt

# Set pandas to display floating-point numbers with more precision
pd.set_option('display.float_format', '{:.10f}'.format)

# Define the calculate_sentiment function
def calculate_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Example setup for headlines
headline_samples = [
    "hits record high",
    "Concerns over future growth",
    "to unveil new product next month",
    "reports earnings that exceed forecasts",
    "Market downturn effects",
    "invests in renewable energy",
    "New CEO announced",
    "faces regulatory scrutiny",
    "rumored to acquire a tech startup",
]

# Define the ticker symbols and time period
tickers = ['AAPL', 'AMZN', 'GOOG', 'FB', 'MSFT', 'NVDA', 'TSLA']
start_date = '2020-01-01'
end_date = '2023-01-01'

# Iterate over each ticker
for ticker in tickers:
    # Download stock data
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    
    # Drop 'Adj Close' if it exists, then rename the remaining 5 columns
    if 'Adj Close' in stock_data.columns:
        stock_data = stock_data.drop(columns=['Adj Close'])
    stock_data.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
    
    # Create a DataFrame with dates from the stock data
    dates = stock_data.index
    text_data = pd.DataFrame({'Date': dates})

    # Assign a random headline to each date (customized with ticker)
    text_data['Headline'] = [f"{ticker} {np.random.choice(headline_samples)}" for _ in range(len(text_data))]

    # Apply sentiment analysis
    text_data['Sentiment'] = text_data['Headline'].apply(calculate_sentiment)

    # Align sentiment scores with the stock closing prices
    combined_data = text_data.set_index('Date').join(stock_data['Close'])

    # Calculate daily returns
    combined_data['Daily_Return'] = combined_data['Close'].pct_change()

    # Display the first few rows of the combined data
    print(f"\nFirst few rows for {ticker}:")
    print(combined_data.head())

    # Create a correlation matrix for Sentiment, Daily_Return, and Close
    correlation_matrix = combined_data[['Sentiment', 'Daily_Return', 'Close']].corr()
    
    # Display the correlation matrix
    print(f"\nCorrelation Matrix for {ticker}:")
    print(correlation_matrix)

    # Visualize the correlation matrix as a heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title(f'Correlation Matrix for {ticker}')
    plt.show()
```
